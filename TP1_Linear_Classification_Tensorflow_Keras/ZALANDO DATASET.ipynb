```python
from tensorflow.keras.layers import *
from tensorflow.keras.activations import *
from tensorflow.keras.losses import *
from tensorflow.keras.models import *
from tensorflow.keras.utils import *
from tensorflow.keras.optimizers import *
from tensorflow.keras.initializers import *
from tensorflow.keras.regularizers import *
from tensorflow.keras.metrics import *
from tensorflow.keras.datasets import *
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import datetime

# Pour les imports, il est mieux d'importer uniquement ce que l'on va utiliser et non faire un * ( wildcard) pour tout importer
```

# Load Dataset

```python
(x_train, y_train), (x_val, y_val) = fashion_mnist.load_data()
```

```python
np.shape(x_train)
```

# One Hot VS Sparse VS 10 Models

```python
y_one_hot_train = tf.one_hot(y_train, 10)
y_one_hot_val = tf.one_hot(y_val, 10)
```

# Reshape 

```python
x_train255 = x_train.reshape(x_train.shape[0], 28*28)
x_val255 = x_val.reshape(x_val.shape[0], 28*28)
```

# Min Max

```python
x_train1 = x_train255 / 255 
x_val1 = x_val255 / 255 
```

# Modèle Linéaire
###   One Hot - Pixel(0, 255)

```python
#training params
epochs = 200
batch_size = 5000

#Create and fit model
model = Sequential()
model.add(Dense(1, activation="linear", input_dim=28*28))
model.add(Dense(10, activation="softmax", input_dim=1))
model.compile(optimizer=Adam(), loss=[categorical_crossentropy], metrics=[categorical_accuracy])
model.fit(x_train255, y_one_hot_train, validation_data=(x_val255, y_one_hot_val), epochs=epochs, batch_size=batch_size)
```

###   One Hot - Pixel(0, 1)

```python
#training params
epochs = 200
batch_size = 5000

#Create and fit model
model = tf.keras.Sequential()
model.add(Dense(1, activation="linear", input_dim=28*28))
model.add(Dense(10, activation="softmax", input_dim=1))
model.compile(optimizer=Adam(), loss=[categorical_crossentropy], metrics=[categorical_accuracy])
model.fit(x_train1, y_one_hot_train, validation_data=(x_val1, y_one_hot_val), epochs=epochs, batch_size=batch_size)
```

###   Sparse - Pixel(0, 255)

```python
#training params
epochs = 1000
batch_size = 5000


#Create and fit model
model = tf.keras.Sequential()
model.add(Dense(1, activation="linear", input_dim=28*28))
model.add(Dense(10, activation="softmax", input_dim=1))
model.compile(optimizer=Adam(), loss=[sparse_categorical_crossentropy], metrics=[sparse_categorical_accuracy])
model.fit(x_train255, y_train, validation_data=(x_val255, y_val), epochs =epochs, batch_size=batch_size)
```

###   Sparse - Pixel(0, 1)

```python
#training params
epochs = 200
batch_size = 5000

#Create and fit model
model = tf.keras.Sequential()
model.add(Dense(1, activation="linear", input_dim=28*28))
model.add(Dense(10, activation="softmax", input_dim=1))
model.compile(optimizer=Adam(), loss=[sparse_categorical_crossentropy], metrics=[sparse_categorical_accuracy])
model.fit(x_train1, y_train, validation_data=(x_val1, y_val), epochs =epochs, batch_size=batch_size)
```

# Conclusion ( Comparaison des modèles )

```python
# Pour les modèles linéaires, les Resultats sont les suivants(pour mes entrainements) pour epochs 200 et batch_size = 5000

# One Hot - Pixel(0, 255) => 20 microsecondes / image
# Model1: loss: 1.3738 - categorical_accuracy: 0.4150 - val_loss: 1.3829 - val_categorical_accuracy: 0.4162
# Model2: loss: 1.4211 - categorical_accuracy: 0.4238 - val_loss: 1.4840 - val_categorical_accuracy: 0.4203
# Model3: loss: 1.5130 - categorical_accuracy: 0.4266 - val_loss: 1.5542 - val_categorical_accuracy: 0.4128

# One Hot - Pixel(0, 1)  => 20 microsecondes / image
# Model1: loss: 1.2768 - categorical_accuracy: 0.5048 - val_loss: 1.2894 - val_categorical_accuracy: 0.4946
# Model2: loss: 1.3095 - categorical_accuracy: 0.4707 - val_loss: 1.3240 - val_categorical_accuracy: 0.4672
# Model3: loss: 1.3468 - categorical_accuracy: 0.4879 - val_loss: 1.3656 - val_categorical_accuracy: 0.4868

# Pour le OneHot avoir des pixels entre 0 et 1 semble avoir de meilleurs résultats

# Sparse - Pixel(0, 255)  => 3 microsecondes / image
# Model1: loss: 1.4526 - sparse_categorical_accuracy: 0.4061 - val_loss: 1.4705 - val_sparse_categorical_accuracy: 0.4036
# Model2: loss: 1.3566 - sparse_categorical_accuracy: 0.4006 - val_loss: 1.3693 - val_sparse_categorical_accuracy: 0.3999
# Model3: loss: 1.5030 - sparse_categorical_accuracy: 0.3744 - val_loss: 1.5537 - val_sparse_categorical_accuracy: 0.3635

# Sparse est beaucoup plus rapide que OneHot pour l'entrainement il prend 3 us par image au lieu de 20 pour le OneHot CEPENDANT
# on remarque que la précision est plus faible (il faut surement tester avec plus d'epochs)

# Testons avec 1000 epochs
# loss: 1.0983 - sparse_categorical_accuracy: 0.5318 - val_loss: 1.1215 - val_sparse_categorical_accuracy: 0.5358
# Verdict: on obtient un meilleur résultat avec sparse en moins de temps d'entrainement.

# Sparse - Pixel(0, 1)  => 13 microsecondes / image
# Model1: loss: 1.2257 - sparse_categorical_accuracy: 0.4769 - val_loss: 1.2384 - val_sparse_categorical_accuracy: 0.4747
# Model2: loss: 1.2105 - sparse_categorical_accuracy: 0.4974 - val_loss: 1.2226 - val_sparse_categorical_accuracy: 0.4942
# Model3: loss: 1.2145 - sparse_categorical_accuracy: 0.4833 - val_loss: 1.2261 - val_sparse_categorical_accuracy: 0.4840

# Sparse avec des pixels entre 0 et 1 est moins rapide que pour les pixels entre 0 et 255 mais reste plus rapide 
# que le oneHot.


###     CONCLUSION 
#
#  Normallement vous êtes maintenant capable de créer et entrainer un modèle linéaire.
#  D'après les différentes résultats il peut etre intéressant de tester les différents modèles pour voir les différences.
```
